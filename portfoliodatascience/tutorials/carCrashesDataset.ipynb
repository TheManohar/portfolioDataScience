```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df =sns.load_dataset('car_crashes')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(3,1,1)
sns.regplot(x=df['total'], y=df['alcohol'])

corr = df.corr()
plt.subplot(3,1,2)
sns.heatmap(corr, annot=True)


dflog = np.log(df.iloc[:,:7])
plt.subplot(3,1,3)
sns.boxplot(data=dflog)

plt.savefig('myplot.svg', dpi=300)
plt.show()

#df.hist()
```
